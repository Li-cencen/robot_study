从DataFrame中选择子集

选择列

```python
ages = titanic["Age"]
print(ages.head())
print(type(ages))
print(ages.shape)
```

```python
age_sex = titanic[["Age", "Sex"]]
print(age_sex.head())
print(type(age_sex))
print(age_sex.shape)
```

选择行

```python
above_35 = titanic[titanic["Age"] > 35]
print(above_35.head())
```

```python
titanic["Age"] > 35
```

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
print(class_23)
```

```python
class_23 == titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
```

```python
age_no_na = titanic[titanic["Age"].notna()]
print(age_no_na)
```

```python
titanic.take([1,3,5])
```

选择指定行列

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
print(adult_names)
```

```python
titanic.iloc[9:25, 2:5]
```

```python
titanic.iloc[0:3, 3] = "anonymous"
print(titanic.head())
```

通过原有数据添加（派生）出新的数据

```python
titanic['Age_now'] = titanic['Age']+10
print(titanic.head())
```

```python
titanic['Age_now'] = np.where(titanic['Survived']==0, np.nan, titanic['Age']+10)
print(titanic.head())
```

DataFrame的描述性统计

```python
titanic["Age"].mean()
```

```python
titanic[["Age", "Fare"]].median()
```

```python
titanic[["Age", "Fare"]].describe()
```

```python
titanic.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
titanic.groupby("Sex").mean(numeric_only=True)
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

```python
titanic["Pclass"].value_counts()
```

```python
titanic.groupby(['Sex',"Pclass"])["Pclass"].count()
```
